##### Assignment 1: Discrete Selector

* Make a Choropleth map of TotalLift in each country
* Allow user to toggle Snowparks and Night Ski Feature.
Your Selector should allow them to filter ski resorts that have one or both these features. if neither is slected the map should sdiplay total lifts in each country

```python
import pandas as pd
import plotly.express as px
from dash import Dash
from dash import html, dcc
from dash.dependencies import Output, Input
import plotly.express as px
import pandas as pd
```

```python
ski_resort = pd.read_csv('/content/European_Ski_Resorts.csv').drop(columns='Unnamed: 0')
ski_resort.head()
```

```python
ski_resort.Country.unique()
```

```python
ski_resort.groupby(['Country'], as_index=False).agg({'LiftCapacity':'sum'})
```

```python
feature='TotalLifts'
px.choropleth(ski_resort.query('px.choropleth(ski_resort.groupby(['Country'], as_index=False).agg({'TotalLifts':'sum'}),
              locationmode='country names',
           scope ='europe', locations='Country',
              color=feature,
              title ="LiftCapacity Distribution across Europe ").update_layout(
                  title = {
                      'x':0.5
                  },
                  paper_bgcolor='white'
              color='TotalLifts',
              title ="LiftCapacity Distribution across Europe ").update_layout(
                  title = {
                      'x':0.5
                  },
                  paper_bgcolor='white'
              )
```

```python
ski_resort.head()
```

```python
app = Dash(__name__)
app.layout = html.Div([
    dcc.Checklist(id = 'feature', options=['Snowparks','NightSki'], value=['Snowparks'],
                  multi=False),
    html.H5(id='title'),
    dcc.Graph(id='expenditure-graph')
])
@app.callback(Output('title','children'),
    Output('expenditure-graph','figure'),
    Input('feature', 'value')
)
def ret_country(feature):
    title=f"{feature}"
    if feature == ["NightSki"]:
        fig = px.choropleth(ski_resort.query("NightSki == 'Yes'").groupby(['Country'], as_index=False).agg({'TotalLifts':'sum'}),
                locationmode='country names',
            scope ='europe', locations='Country',
                color='TotalLifts',
                title ="LiftCapacity Distribution across Europe ").update_layout(
                    title = {
                        'x':0.5
                    },
                    paper_bgcolor='white'
                )

    elif feature==['Snowparks']:
        fig = px.choropleth(ski_resort.query('Snowparks == "Yes"').groupby(['Country'], as_index=False).agg({'TotalLifts':'sum'}),
                locationmode='country names',
            scope ='europe', locations='Country',
                color='TotalLifts',
                title ="LiftCapacity Distribution across Europe ").update_layout(
                    title = {
                        'x':0.5
                    },
                    paper_bgcolor='white'
                )
    elif len(feature) ==2:
        fig = px.choropleth(ski_resort.query('Snowparks == "Yes" and NightSki == "Yes"').groupby(['Country'], as_index=False).agg({'TotalLifts':'sum'}),
                locationmode='country names',
            scope ='europe', locations='Country',
                color='TotalLifts',
                title ="LiftCapacity Distribution across Europe ").update_layout(
                    title = {
                        'x':0.5
                    },
                    paper_bgcolor='white'
                )

    else:
        fig = px.choropleth(ski_resort.groupby(['Country'], as_index=False).agg({'TotalLifts':'sum'}),
                locationmode='country names',
            scope ='europe', locations='Country',
                color='TotalLifts',
                title ="LiftCapacity Distribution across Europe ").update_layout(
                    title = {
                        'x':0.5
                    },
                    paper_bgcolor='white'
                )

    return  title,fig



if __name__ == '__main__':
    app.run(debug=True)
```
